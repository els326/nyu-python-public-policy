Step 4. Get the count of 311 complaints per community district (borocd) using the filtered dataframe you created in Step 2.

```
traffic_count = df_311.groupby('borocd').size().reset_index(name='count_traffic_signal')
print(traffic_count)
```

Step 5. Join the indicator dataframe (from Step 1) with 311 counts dataframe (from Step 4) using the borocd value.

```
df_merged = pd.merge(how='left', left=traffic_count, right=df_commute, left_on='borocd', right_on = 'borocd')
df_merged.head()
```

Step 6. Produce a scatter plot of the indicator values vs. 311 complaint counts. The should be one point per community district.

```
x = np.array(df_merged['mean_commute'])
y = np.array(df_merged['count_traffic_signal'])

color = '#a83232' 
area = 50 

plt.figure(figsize=(15, 6))

plt.scatter(x, y, s=area, c=color, edgecolors='black', alpha=0.5) 

plt.title('Traffic Signal Complaints vs. Mean Commute')
plt.xlabel('Mean Commute')
plt.ylabel('# of Traffic Signal Complaints in Community District')

plt.show()
```

Step 7. Conduct an OLS linear regression using the indicator values vs. 311 complaint counts and interpret the results.

```
import statsmodels.api as sm
```

```
x = np.array(df_merged['mean_commute'])
y = np.array(df_merged['count_traffic_signal'])

np.corrcoef(x, y)
```

```
y = np.array(df_merged['mean_commute'])
x = np.array(df_merged['count_traffic_signal'])
x = sm.add_constant(x)
print(x)
model = sm.OLS(y,x).fit()
print(model.summary())
```

The relationship between traffic signal complaints and mean commute time is very weak. The R squared is .047, which tells us that only 4.7% of the variation in mean commute time is explained by traffic signal problems. The P value also indicates that the relationship between the two variables is not statistically significant.

Step 8. Use the linear regression results to add a line of best fit to your scatterplot

```
p = model.params 
print('parameters:', p)

r2 = model.rsquared
print('R2:', r2)
```

```
label = 'y = ' + str(round(p[1],2)) + 'x + ' + str(round(p[0],2)) + '; R2 = ' + str(round(r2,3))
print(label)
```

```
x_dummy = np.array(df_merged['mean_commute'])

plt.figure(figsize=(15, 6))

plt.scatter(
    np.array(df_merged['mean_commute']), 
    np.array(df_merged['count_traffic_signal']), 
    s=50, 
    c='#3d03fc',
    edgecolors='black', 
    alpha=0.5) 

plt.plot(
    x_dummy, 
    p[0] + p[1] * x_dummy,
    c='#00d138',
    label=label)

plt.title('Traffic Signal Complaints vs. Mean Commute Time')
plt.xlabel('Mean Commute')
plt.ylabel('# of Traffic Signal Complaints')
plt.legend()

plt.show()
```
